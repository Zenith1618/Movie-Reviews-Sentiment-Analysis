```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

#### Define vocabulary size

```python
voc_size = 10000
```

### One hot Representation

```python
one_hot_rep = [one_hot(word, voc_size) for word in sent]
```

```python
one_hot_rep
```

Look at each word index

```python
sent
```

### Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential

import numpy as np
```

We need to make all the sentences of equal size for RNN

```python
sent_length = 8

# Padding
embedded_docs = pad_sequences(one_hot_rep, padding='pre', maxlen=sent_length)   # you can do pre padding(adding zeroes front) or post padding(back)
embedded_docs
```

#### Feature Representation

```python
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
